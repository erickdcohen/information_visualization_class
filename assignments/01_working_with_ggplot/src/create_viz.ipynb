# Assignment 01 Minard Visualization GGPlot


```python
import pandas as pd
from plotnine import (ggplot, aes, geom_path, geom_text, geom_line, theme,
                      scale_size, scale_color_manual, xlab, ylab, lims, save_as_pdf_pages)

# path constants text files
PATH_TO_CITIES_TXT = "../data/cities.txt"
PATH_TO_TEMPS_TXT = "../data/temps.txt"
PATH_TO_TROOPS_TXT = "../data/troops.txt"

# read in the files
cities_df = pd.read_csv(PATH_TO_CITIES_TXT, sep=r'\s+')
temps_df = pd.read_csv(PATH_TO_TEMPS_TXT, sep=r'\s+')
troops_df = pd.read_csv(PATH_TO_TROOPS_TXT, sep=r'\s+')
```

```python
plots_troops = (
    ggplot(troops_df, aes("long", "lat")) +
    geom_path(aes(size="survivors", color="direction",
              group="group"), lineend="round")
)

plot_both = (
    plots_troops +
    geom_text(aes(label="city"), size=14, data=cities_df)
)

plot_polished = (
    plot_both +
    scale_size(
        (1, 20),
        breaks=list(map(lambda x: x * 10 ^ 5, (1, 2, 3))),
        labels=list(map(lambda x: x * 10 ^ 5, (1, 2, 3))),
    ) +
    scale_color_manual(values=("#9e9e9e", "#f30f0f")) +
    xlab(None) +
    ylab(None) +
    theme(figure_size=(15, 6))
)

plot_polished
```

```python
plot_temps = (
    ggplot(temps_df, aes("long", "temp")) +
    geom_line() +
    geom_text(aes(label="month"), nudge_x=-0.55, nudge_y=0.2, size=14) +
    geom_text(aes(label="day"), nudge_x=-0.1, nudge_y=0.2, size=14) +
    theme(figure_size=(15, 6)) +
    lims(x=(23, 40))
)

plot_temps
```

```python
plot_polished.save("../output/minard_troops.png")
plot_temps.save("../output/minard_temps.png")
```
